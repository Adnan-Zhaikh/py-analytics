```python
import pandas as pd
import matplotlib.pyplot as plt

turnaround_times = [12,14,15,14,16,15,12,13,14,15,90,15,95]
df = pd.DataFrame(turnaround_times, columns = ["Turnaround"])
```

```python
fig, ax = plt.subplots()

ax.boxplot(df["Turnaround"], orientation="horizontal")
ax.set_title("Operational Turnaround Dispersion")
ax.set_xlabel("Hours")
plt.show()
```

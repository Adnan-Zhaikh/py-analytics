```python
import pandas as pd
import numpy as np

# A dirty dataset simulating missing warehouse records
dirty_data = {
    "Product": ["Laptop", "Mouse", "Monitor", "Keyboard", "Headset"],
    "UnitsSold": [120, np.nan, 85, np.nan, 50],
    "Revenue": [90000, 1500, np.nan, 1200, 4000]
}
df = pd.DataFrame(dirty_data)
print("--- Original Dirty DataFrame ---")
print(df)
print("\n")
```

```python
dropped_df = df.dropna()
print(dropped_df)

print("--------------")
filled_df = df.fillna(0)
print(filled_df)
```

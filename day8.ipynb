```python
import pandas as pd

corporate_data = {
    "Team": ["Engineering", "Sales", "Engineering", "Sales", "Design", "Engineering"],
    "Employee": ["Adnan", "Sarah", "Billa", "John", "Tara", "Sahi"],
    "MonthlyRevenue": [15000, 22000, 18000, 19000, 12000, 16000]
}
df = pd.DataFrame(corporate_data)
print("--- Raw Corporate Records ---\n", df, "\n")
```

```python
revenue_by_team = df.groupby("Team")["MonthlyRevenue"].sum()
print("Total Revenue by Team: ", revenue_by_team)

average_revenue_by_team = df.groupby("Team")["MonthlyRevenue"].mean()
print("Average Revenue by Team: ", average_revenue_by_team)
```

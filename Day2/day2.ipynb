```python
product_catalog = ["Mouse","Keyboard","Monitor","USB-C Cable"]
stock_counts = [5,40,0,50]
unit_cost = [245.0,150.0,550.0,12.0]
print("\n---Initializing Automated warehouse Stock Audit---")

for i in range(len(product_catalog)):
    product = product_catalog[i]
    stock = stock_counts[i]
    cost = unit_cost[i]
    total_value = stock * cost
    print(f"Stock {i} | {product} -> Stock: {stock} units | Asset value: ${total_value:,.2f}")
```

```python
out_of_stock = []
critical_reorder_list = []
stable_stock_list = []

for i in range(len(product_catalog)):
    product = product_catalog[i]
    stock = stock_counts[i]

    if stock == 0:
        out_of_stock.append(product)
    elif stock <= 15:
        critical_reorder_list.append({"item": product, "current_stock": stock})
    else:
        stable_stock_list.append(product)

print("\n----Logistics analytics reports----")
print("CRITICAL SQS (Out of Stock): ", out_of_stock)
print("REORDER IMMINENT (Stock<=15): ", critical_reorder_list)
print("STABLE STOCKS (Safe zone): ", stable_stock_list)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Days": ["Day 1", "Day 2", "Day 3"],
    "BugsFixed": [4, 14, 10],
    "HoursWorked": [6,5,8]
}

df = pd.DataFrame(data)
```

```python
fig, ax = plt.subplots(1, 2)
ax[0].plot(df["Days"], df["BugsFixed"], color = "red")
ax[0].set_title("Bugs Flattened")
ax[1].bar(df["Days"], df["HoursWorked"], color="green")
ax[1].set_title("Time Investment")

plt.tight_layout()
plt.show()
```

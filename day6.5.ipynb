```python
import pandas as pd
raw_data = {
    "Employee": ["Adnan", "Addy", "Billa", "Sahi"],
    "Department": ["Engineering", "Design", "Data Science", "Engineering"],
    "ProjectsCompleted": [12, 5, 9, 7],
    "IsActive": [True, True, False, True]
}
df = pd.DataFrame(raw_data)
print(df)
```

```python
active_team = df[df['IsActive'] == True]

print("--- Active Team Members ---")
print(active_team)
```

```python
elite_engineers = df[(df["Department"] == "Engineering") & (df["ProjectsCompleted"]>5)]

print("--- Elite Engineer ---")
print(elite_engineers)
```

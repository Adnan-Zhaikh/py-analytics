```python
import pandas as pd 

# The raw log tracking metrics
sales_data = {
    "Year": [2025, 2025, 2026, 2026],
    "Region": ["North", "South", "North", "South"],
    "Revenue": [8000, 6000, 9500, 7500]
}
df = pd.DataFrame(sales_data)
```

```python
pivot_df = df.pivot(index="Year", columns="Region", values="Revenue")

print(pivot_df)
```

```python
import pandas as pd 
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8-whitegrid')

quarter = ["Q1", "Q2", "Q3", "Q4"]
converstions = [450, 590, 840, 700]
df = pd.DataFrame({"Quarter": quarter, "Conversions": converstions})
```

```python
fig, ax = plt.subplots(figsize=(7, 4))
ax.bar(df["Quarter"], df["Conversions"], color="#4CAF50", width=0.6)
ax.grid(True, linestyle="--", alpha=0.6, color="gray")
ax.set_title("Quarterly Conversion Performance")

plt.show()
```

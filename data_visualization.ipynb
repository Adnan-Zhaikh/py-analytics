```python
import pandas as pd
import matplotlib.pyplot as plt

metrics = {
    "Days": ["Day 1", "Day 2", "Day 3", "Day 4"],
    "Projects": [3, 5, 9, 6]
}
df = pd.DataFrame(metrics)
```

```python
fig, ax = plt.subplots()
ax.plot(df["Days"], df["Projects"], marker="o", color="blue")

ax.set_title("Velocity Tracker")
ax.set_ylabel("Projects Completed")

plt.show()
```

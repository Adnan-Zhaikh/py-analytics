```python
data_store = {
    "order_id": [101,102,103,104],
    "item": ["Sublimation keychain", "UV printed plate","Corporate mug", "Custom medal"],
    "unit_price": [50.0, 180.0, 150.0, 85.0],
    "quantity": [50, 10, 0, 5]
}
print("Raw data Map:\n", data_store)
```

```python
items = data_store["item"]
quantities = data_store["quantity"]
print("\n-------Scannning Data for Empty Orders-------")
for i in range(len(items)):
    current_item = items[i]
    current_qty = quantities[i]

    if current_qty == 0:
        print(f"ALERT: Order for '{current_item}' has 0 quantity. Flagging record")
    else:
        print(f"Record clear: '{current_item}' contains valid quantiity of {current_qty}")
```

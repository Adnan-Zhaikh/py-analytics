```python
import pandas as pd
import matplotlib.pyplot as plt

# A tracking list of project hours across dozens of sprints
sprint_hours = [14, 15, 16, 14, 22, 23, 15, 16, 25, 24, 15, 16, 14, 28, 15, 16, 22, 23]
df = pd.DataFrame(sprint_hours, columns=["Hours"])
```

```python
fig, ax = plt.subplots()
ax.hist(df["Hours"], bins=5, color="purple", edgecolor="black")
ax.set_title("Sprint Complexity Distribution")
ax.set_xlabel("Hours Expended")
ax.set_ylabel("Sprint Frequency Count")
plt.show()
```
